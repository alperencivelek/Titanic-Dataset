```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("train.csv")
df.info()
df.describe()
```

```python
yas_ortalamasi=round(df['Age'].mean())
print(yas_ortalamasi)
```

```python
#yasi bilinmeyenleri ortalama ile dolduralim
df['Age'].fillna(value=yas_ortalamasi, inplace=True)
#Yaşlari gruplayalım
df.loc[ df['Age'] <= 16, 'Age'] = 0
df.loc[(df['Age'] > 16) & (df['Age'] <= 32), 'Age'] = 1
df.loc[(df['Age'] > 32) & (df['Age'] <= 48), 'Age'] = 2
df.loc[(df['Age'] > 48) & (df['Age'] <= 64), 'Age'] = 3
df.loc[ df['Age'] > 64, 'Age'] = 4
```

```python
##2 null değerli embarked sütuun yolcuların %70'i Southampton S bindiği için S  ile dolduruyorum
df['Embarked'].fillna("S", inplace=True)
df.info()
```

```python
##Çok null değerlere sahip olduğu için cabin sütunun siliyorum
df=df.drop('Cabin',axis=1)
df
```

```python
##Embarked ve Yaş sütununu encode edelim
df=pd.get_dummies(df, columns=['Embarked','Age'])
df
```

```python
##Cinsiyeti binary değişken yapalım
df['Sex'] = df['Sex'].map({'female': 1, 'male': 0}).astype(int)
df
```

```python
#Güçlü bir ilişki olduğunu düşünmediğim için İsim Bilet ve Ücret bilgisini siliyorum
df=df.drop(['Name','Ticket','Fare'],axis=1)
df
```

```python
bagimsiz_Degiskenler=df.iloc[:,2:].values### X değerlerimiz
tahmin_edilecek=df.iloc[:,1].values### y değerimiz
```

```python
##Aynı değişiklikleri test dataseti üzerinde de yapıyorum
test_df=pd.read_csv("test.csv")
yas_ortalamasi=round(test_df['Age'].mean())
print(yas_ortalamasi)
test_df=test_df.drop(['Name','Ticket','Fare','Cabin'],axis=1)
test_df['Age'].fillna(value=yas_ortalamasi, inplace=True)
test_df['Sex'] = test_df['Sex'].map({'female': 1, 'male': 0}).astype(int)
test_df['Embarked'].fillna("S", inplace=True)
test_df.loc[ test_df['Age'] <= 16, 'Age'] = 0
test_df.loc[(test_df['Age'] > 16) & (test_df['Age'] <= 32), 'Age'] = 1
test_df.loc[(test_df['Age'] > 32) & (test_df['Age'] <= 48), 'Age'] = 2
test_df.loc[(test_df['Age'] > 48) & (test_df['Age'] <= 64), 'Age'] = 3
test_df.loc[ test_df['Age'] > 64, 'Age'] = 4
test_df=pd.get_dummies(test_df, columns=['Embarked','Age'])
X_test  = test_df.drop("PassengerId", axis=1).copy()
```

```python
from sklearn.linear_model import LogisticRegression  
classifier= LogisticRegression(random_state=0)  
classifier.fit(bagimsiz_Degiskenler, tahmin_edilecek)
Y_pred = classifier.predict(X_test)
acc_log = round(classifier.score(bagimsiz_Degiskenler, tahmin_edilecek) * 100, 2)
acc_log
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(bagimsiz_Degiskenler,tahmin_edilecek)
Y_pred = svc.predict(X_test)
acc_log = round(svc.score(bagimsiz_Degiskenler, tahmin_edilecek) * 100, 2)
acc_log
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(bagimsiz_Degiskenler,tahmin_edilecek)
Y_pred = knn.predict(X_test)
acc_log = round(knn.score(bagimsiz_Degiskenler, tahmin_edilecek) * 100, 2)
acc_log
```

```python
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier()
RF.fit(bagimsiz_Degiskenler,tahmin_edilecek)
Y_pred = RF.predict(X_test)
acc_log = round(RF.score(bagimsiz_Degiskenler, tahmin_edilecek) * 100, 2)
acc_log
```

```python
from sklearn.tree import DecisionTreeClassifier
DecisionTree=DecisionTreeClassifier()
DecisionTree.fit(bagimsiz_Degiskenler,tahmin_edilecek)
Y_pred = DecisionTree.predict(X_test)
acc_log = round(DecisionTree.score(bagimsiz_Degiskenler, tahmin_edilecek) * 100, 2)
acc_log
```
